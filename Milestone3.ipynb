```python
from pymongo import UpdateOne

@step
def data_ingestion() -> pd.DataFrame:
    print("Ingesting...")
    df = pd.read_csv("substitutes.csv")

    if not df.empty:
        records = df.to_dict(orient="records")
        bulk_operations = []
        for record in records:
            bulk_operations.append(
                UpdateOne(
                    {"Ingredient": record["Ingredient"]},
                    {"$setOnInsert": record},
                    upsert=True
                )
            )

        # Perform the bulk upsert operation
        if bulk_operations:
            collection.bulk_write(bulk_operations)

    # Retrieve all data from MongoDB
    data = pd.DataFrame(list(collection.find({}, {"_id": 0})))
    return data
```

```python
import uuid
from datetime import datetime

@step
def data_preprocessing(df: pd.DataFrame) -> pd.DataFrame:
    print("Preprocessing ingredient data...")
    def clean_text(text):
        text = str(text).lower().strip()
        text = re.sub(r"\s+", " ", text)
        return text

    if "Ingredient" in df.columns:
        df["Ingredient"] = df["Ingredient"].apply(clean_text)

    if "Substitute" in df.columns:
        df["Substitute"] = df["Substitute"].apply(clean_text)

    if "Context" in df.columns:
        df["Context"] = df["Context"].apply(clean_text)

    if "Equivalent" in df.columns:
        df["Equivalent"] = df["Equivalent"].apply(clean_text)

    df["event_timestamp"] = datetime.utcnow()
    df = df.dropna()

    def parse_equivalent(eq):
        match = re.match(r"([\d/.]+)\s*([\w]+)", eq)
        if match:
            quantity = eval(match.group(1))
            unit = match.group(2)
            return quantity, unit
        return None, None

    df["Equivalent Quantity"], df["Equivalent Unit"] = zip(*df["Equivalent"].apply(parse_equivalent))

    print("Generating ingredient embeddings...")
    df["Ingredient Embedding"] = df["Ingredient"].apply(lambda x: nlp.encode(x).tolist())
    df["Substitute Embedding"] = df["Substitute"].apply(lambda x: nlp.encode(x).tolist())

    df["ingredient_id"] = df["Ingredient"].apply(lambda x: str(uuid.uuid5(uuid.NAMESPACE_DNS, x)))
    df["substitute_id"] = df["Substitute"].apply(lambda x: str(uuid.uuid5(uuid.NAMESPACE_DNS, x)))

    print("Preprocessing completed.")
    return df
```

```python
@step
def data_validation(df: pd.DataFrame) -> pd.DataFrame:
    print("Validating data...")
    required_columns = ["Ingredient", "Substitute", "Context", "Equivalent",
                        "Equivalent Quantity", "Equivalent Unit"]
    missing_columns = [col for col in required_columns if col not in df.columns]
    if missing_columns:
        raise ValueError(f"Missing columns: {missing_columns}")

    missing_values = df.isnull().sum()
    if missing_values.sum() > 0:
        print(f"Missing values detected:\n{missing_values[missing_values > 0]}")

    invalid_substitutes = df["Substitute"].apply(lambda x: not isinstance(x, str)).sum()
    if invalid_substitutes > 0:
        print(f"Invalid 'Substitute' formats: {invalid_substitutes}")

    invalid_equivalents = df["Equivalent Quantity"].apply(lambda x: not isinstance(x, (int, float))).sum()
    if invalid_equivalents > 0:
        print(f"Invalid 'Equivalent Quantity' values: {invalid_equivalents}")

    print("Data validation complete.")
    return df
```

```python
@step
def data_versioning(df: pd.DataFrame):
    print("Starting data versioning with DVC...")
    versioning_dir = "/content/dvc_data/"
    os.makedirs(versioning_dir, exist_ok=True)
    timestamp = time.strftime("%Y%m%d-%H%M%S")
    versioned_filename = os.path.join(versioning_dir, f"processed_data_{timestamp}.csv")
    df.to_csv(versioned_filename, index=False)
    time.sleep(2)

    if not os.path.exists(versioned_filename):
        raise FileNotFoundError(f"Error: File {versioned_filename} not found after saving.")

    try:
        print("Performing DVC versioning...")
        subprocess.run(["dvc", "add", versioned_filename], check=True)
        subprocess.run(["git", "add", versioned_filename, f"{versioned_filename}.dvc"], check=True)
        subprocess.run(["git", "commit", "-m", f"Versioned dataset: {timestamp}"], check=True)
        print("DVC PUSH...")
        subprocess.run(["dvc", "push"], check=True)

        print(f"Data saved as {versioned_filename} and successfully versioned with DVC.")

    except subprocess.CalledProcessError as e:
        print(f"DVC versioning failed: {e}")
```

```python
import os
import yaml
import subprocess

def initialize_laddaty_feature_store():
    feature_repo_path = "/content/laddaty_feature_repo"
    feature_store_yaml = os.path.join(feature_repo_path, "feature_store.yaml")

    if not os.path.exists(feature_repo_path):
        print(f"Creating Laddaty feature store at {feature_repo_path}...")
        os.makedirs(feature_repo_path)

        # Install and Start Redis (Online Store)
        redis_status = subprocess.run(["redis-server", "--version"], capture_output=True, text=True)
        if redis_status.returncode != 0 or "Redis server" not in redis_status.stdout:
            print("Installing Redis...")
            subprocess.run(["apt-get", "update", "-y"])
            subprocess.run(["apt-get", "install", "redis-server", "-y"])
            print("Starting Redis server...")
            subprocess.run(["redis-server", "--daemonize", "yes"])

        # Feature Store Configuration
        feature_store_config = {
            "project": "laddaty_ingredient_store",
            "registry": os.path.join(feature_repo_path, "registry.db"),
            "provider": "local",
            "online_store": {
                "type": "redis",
            },
            "offline_store": {
                "type": "file",
            },
            "entity_key_serialization_version": 2
        }

        with open(feature_store_yaml, "w") as f:
            yaml.dump(feature_store_config, f)
        print("Feature store configuration created.")
    else:
        print(f"Laddaty feature store already exists at {feature_repo_path}.")

initialize_laddaty_feature_store()
```

```python
import uuid
from feast import FeatureStore, Entity, Field, FeatureView, FileSource, ValueType
from feast.types import Array, Float32, String

@step
def feature_store_integration(df: pd.DataFrame):
    print("Updating Feature Store...")
    feature_repo_path = os.path.abspath("/content/laddaty_feature_repo")
    os.makedirs(feature_repo_path, exist_ok=True)
    parquet_path = os.path.join(feature_repo_path, "ingredient_features.parquet")
    df.to_parquet(parquet_path, index=False)
    ingredient_entity = Entity(name="ingredient_id", value_type=ValueType.STRING, description="Unique Ingredient ID")

    ingredient_embedding_feature = Field(name="Ingredient Embedding", dtype=Array(Float32))
    substitute_feature = Field(name="substitute_id", dtype=String)
    equivalent_quantity_feature = Field(name="equivalent_quantity", dtype=Float32)
    equivalent_unit_feature = Field(name="equivalent_unit", dtype=String)
    batch_source = FileSource(
        path=os.path.abspath(parquet_path),
        timestamp_field="event_timestamp"
    )

    ingredient_feature_view = FeatureView(
        name="ingredient_features",
        entities=[ingredient_entity],
        schema=[
            ingredient_embedding_feature, substitute_feature,
            equivalent_quantity_feature, equivalent_unit_feature
        ],
        source=batch_source
    )
    feature_store_yaml = os.path.join(feature_repo_path, "feature_store.yaml")
    if not os.path.exists(feature_store_yaml):
        feature_store_config = {
            "project": "laddaty_feature_store",
            "registry": os.path.join(feature_repo_path, "registry.db"),
            "provider": "local",
            "offline_store": {"type": "file"},
            "online_store": {"type": "sqlite", "path": os.path.join(feature_repo_path, "online_store.db")}
        }
        with open(feature_store_yaml, "w") as f:
            yaml.dump(feature_store_config, f)
        print("Created feature_store.yaml configuration.")

    try:
        feature_repo_path = os.path.abspath("laddaty_feature_repo")
        store = FeatureStore(repo_path=feature_repo_path)
        store.apply([ingredient_feature_view])
        print("Features stored successfully in Feast.")
    except Exception as e:
        print(f"Error while updating Feast feature store: {e}")
```

```python
@pipeline(enable_cache=False)
def data_pipeline():
    df = data_ingestion()
    df = data_preprocessing(df)
    df = data_validation(df)
    data_versioning(df)
    feature_store_integration(df)

data_pipeline()
```
